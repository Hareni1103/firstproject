```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/NM Weather Data in India from 1901 to 2017 - Weather Data in India from 1901 to 2017.csv")
df
```

```python
df.columns = ['ID','YEAR','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
df.head(5)
```

```python
corr=df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f')
plt.title("Correlation Matrix")
```

## Visualizing the monthly Temperature

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Month',y='Temperature',data=pd.melt(
    df,id_vars=['ID','YEAR'],var_name='Month',value_name='Temperature')) #Pandas melt was used to transform the dataset from a Wide format into a Long format.
plt.title("Distribution of Temperatures")
```

**January starts with a relatively low temperature of approximately 18 degrees. The temperature gradually increases as we move through the months, reaching a peak in May at greater than 28 degrees. Subsequently, it begins to decrease slowly, reaching around 19 degrees by December.**



## Visualizing the Yearly temperature

```python
plt.figure(figsize=(15,8))
selected_years=range(2008,2018)
selected_data=df[df['YEAR'].isin(selected_years)]   #filtering the dataframe
selected_data['YEAR']=selected_data['YEAR'].astype(int)   #converting the year column to int
selected_data=selected_data.drop(columns=['ID'])
average_temps=selected_data.groupby('YEAR').mean()   #avg temps for each year
for year in selected_years:
    plt.plot(average_temps.columns,average_temps.loc[year,:],marker='x',label=str(year))

plt.title("Average Monthly Temperature for the Last 10 Years")
plt.xlabel("Month")
plt.ylabel("Average Temperature")
plt.legend()
```

**The dataset shows a trend where the Earth's temperature tends to increase over the years, indicating a pattern of global warming. For instance, in the summer month of May, the data suggests that in the year 2008, the temperature was around 29 degrees, but in 2018, it had risen to 31 degrees. This observation aligns with the broader trend of rising temperatures, which is a concerning aspect associated with global warming.**
